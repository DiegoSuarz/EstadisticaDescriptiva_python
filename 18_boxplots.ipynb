## Diagrama de caja y bigotes

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
np.random.seed(19880519)
```

```python
dist1 = 100*np.random.rand(50)
dist2 = 50*np.ones(25)
dist3 = 100+100*np.random.rand(10)
dist4 = -100*np.random.rand(10)
data = np.concatenate((dist1, dist2, dist3, dist4))
data
```

```python
fig1, ax1 = plt.subplots()
ax1.set_title("Boxplot básico")
ax1.boxplot(data)
```

```python
fig2, ax2 = plt.subplots()
ax2.set_title("Boxplot básico")
ax2.boxplot(data, notch= True)
```

```python
green_diamond = dict(markerfacecolor = 'g', marker= 'D')
fig3, ax3 = plt.subplots()
ax3.set_title("Boxplot básico")
ax3.boxplot(data, flierprops= green_diamond)
```

```python
fig4, ax4 = plt.subplots()
ax4.set_title("Boxplot sin outliers")
ax4.boxplot(data, showfliers = False)
```

```python
red_square = dict(markerfacecolor = "r", marker = "s")
fig5, ax5 = plt.subplots()
ax5.set_title("Boxplot personalizado y en horizontal")
ax5.boxplot(data, vert= False, flierprops = red_square)
```

```python
red_square = dict(markerfacecolor = "r", marker = "s")
fig5, ax5 = plt.subplots()
ax5.set_title("Boxplot personalizado y en horizontal")
ax5.boxplot(data,  # datos
            vert= False, # formato horizontal
            flierprops = red_square, # rojo, cuadrado
            whis = 0.75 # cambiar el tamaño del rango intercuartilico (cambia el tamaño de los bigotes) ,por defecto es 1.5
            ) 
```
