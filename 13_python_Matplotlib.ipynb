## Multiples gráficos en una misma figura:

```python
def f(x):
    return np.exp(-x)*np.cos(2*np.pi*x)
```

```python
x1 = np.arange(0, 5, 0.1)
x2 = np.arange(0, 5, 0.2)

plt.figure(1) #crear una figura
plt.subplot(211) #2 filas, 1 columna, 1er grafico , plt.subplot(211) = plt.subplot(2,1,1)
plt.plot(x1, f(x1), "ro", x2, f(x2), "k") #rojo, o = circ

plt.subplot(212) #2 filas, 1 columna, 2do grafico
plt.plot(x2, f(x2), 'g--') #verde, -- = lineas punteadas

plt.show()
```

```python
plt.figure(1) #crear una figura
plt.subplot(2,1,1) #2 filas, 1 columna, 1er grafico
plt.plot([1,2,3])
plt.subplot(2,1,2) #2 filas, 1 columna, 2do grafico
plt.plot([4,5,6])

plt.figure(2)
plt.plot([4,5,6])

plt.figure(1)
plt.subplot(2,1,1)
plt.title("Esto es el primer titulo")
```

## Añadir texto a los gráficos

```python
mu = 100 # media
sigma = 20 #  desviacion estandar

x = mu + sigma*np.random.randn(10000) #datos aleatorios
```

```python
n , bins, patches = plt.hist(x, 50, density=True, facecolor="g", alpha=0.6) #histograma
plt.xlabel("Cociente Intelectual", fontsize=14, color="green") 
plt.ylabel("Probabilidad")
plt.title(  r"Histograma de CI $N(\mu, \sigma)$")
plt.text(40, 0.015, r'$\mu = 100, \ \sigma=20$')  #coordenadas x, y, texto
plt.axis([20,180,0, 0.025]) #eje x de 20 a 180, eje y de 0 a 0.025
plt.grid()
plt.show()
```

```python
plt.figure(figsize=(10,6), dpi=80) #tamaño de la figura, dpi = pixeles por pulgada
plt.subplot(1,1,1)

x = np.arange(0, 10*np.pi, 0.01)
y = np.cos(x)

plt.plot(x,y, lw=2.0) #grosor de la linea

plt.annotate("Máximo Local",  #texto
            xytext=(15, 1.5), # coordenadas del texto
            arrowprops=dict(facecolor="black", shrink=0.05) ,#flecha
            xy=(4*np.pi, 1)) # coordenadas de la flecha         
            
plt.ylim(-2,2) #eje y de -2 a 2
plt.show()
```

## Cambios de escala y ejes

```python
from matplotlib.ticker import NullFormatter #importar NullFormatter

mu = 0.5 #media
sd = 0.3 #desviacion estandar
y = mu + sd*np.random.randn(1000) #datos aleatorios
y = y[(y>0) & (y<1)] #filtrar datos
y.sort() #ordenar datos de menor a mayor
x = np.arange(len(y)) #crear un array de 0 a len(y)

plt.figure(figsize=(10,10), dpi=80)
plt.subplot(2,2,1) #2 filas, 2 columnas, 1er grafico
plt.plot(x,y)
plt.yscale("linear") #escala lineal
plt.xscale("linear") #escala lineal
plt.title("#scala Lineal")
plt.grid(True)

plt.subplot(2,2,2) #2 filas, 2 columnas, 2do grafico
plt.plot(x, y)
plt.yscale("log") #escala logaritmica
plt.title("Escala Logaritmica")
plt.grid(True)

plt.subplot(2,2,3) #2 filas, 2 columnas, 3er grafico
plt.plot(x, y - y.mean())
plt.yscale("symlog", linthresh=0.01) #escala simetrica
plt.title("Escala Simetrica")
plt.grid(True)

plt.subplot(2,2,4) #2 filas, 2 columnas, 4to grafico
plt.plot(x, y)
plt.yscale("logit") #escala logistica
plt.title("Escala Logistica")
plt.gca().yaxis.set_minor_formatter(NullFormatter()) #eliminar los numeros menor
plt.grid(True)

plt.subplots_adjust(top=0.92, bottom=0.08, left=0.10, right=0.95, hspace=0.25, wspace=0.35) #ajustar los subgraficos
plt.show()
```

## Cambios en los ejes

```python
x = np.linspace(-np.pi, np.pi, 256, endpoint=True) #de -pi a pi, 256 puntos,  incluir el punto final
S, C = np.sin(x), np.cos(x) #seno y coseno

plt.plot(x, S, color = "blue", linewidth = 1.2, linestyle = "-", label = "seno") #azul, linea continua, etiqueta = seno
plt.plot(x, C, color = "green", linewidth = 1.2, linestyle = "-", label = "coseno") #verde, linea continua, etiqueta = coseno
plt.xlim(-4,4) #eje x de -4 a 4
#plt.ylim(-1.2,1.2) #eje y de -1.2 a 1.2
plt.ylim(S.min()*1.1, S.max()*1.1) #eje y de S.min()*1.1 a S.max()*1.1
#plt.xticks(np.linspace(-4,4,9, endpoint=True)) #etiquetas del eje x, de -4 a 4, 9 puntos
plt.xticks([-np.pi, -np.pi/2, 0, np.pi/2, np.pi], [r'$-\pi$', r'$-\pi/2$', r'$0$', r'$\pi/2$', r'$\pi$']) #etiquetas personalizadas
#plt.yticks(np.linspace(-1,1,9,endpoint=True)) #etiquetas del eje y, de -1 a 1, 9 puntos
plt.yticks(np.linspace(-1,1,3, endpoint=True), [-1, " ", 1]) #etiquetas personalizadas

ax = plt.gca() #obtener los ejes

ax.spines["right"].set_color("none")    #eliminar borde derecho
ax.spines["top"].set_color("none")    #eliminar borde superior

ax.xaxis.set_ticks_position("bottom")   #posicion de las marcas del eje x
ax.spines["bottom"].set_position(("data",0))    #posicion del eje x

ax.yaxis.set_ticks_position("left")   #posicion de las marcas del eje y
ax.spines["left"].set_position(("data",0))    #posicion del eje y

plt.legend(loc="upper left") #mostrar la leyenda en la esquina superior izquierda

x0 = 2*np.pi/3
plt.plot([x0, x0], [0, np.sin(x0)], color="red", linewidth=2.5, linestyle="--") #linea vertical

plt.scatter([x0, ], [np.sin(x0), ], 50, color="red") #punto rojo

plt.annotate(r"$\sin(\frac{2\pi}{3}) = \frac{\sqrt{3}}{2}$", 
            xy = (x0, np.sin(x0)), xycoords= "data", 
            xytext = (+10,+30), textcoords="offset points",
            fontsize = 16,
            arrowprops=dict(arrowstyle = "->",connectionstyle="arc3, rad=.2"))   #texto

for label in ax.get_xticklabels() + ax.get_yticklabels():
    label.set_fontsize(16)
    label.set_bbox(dict(facecolor="white", edgecolor="None", alpha=.7)) #etiquetas

plt.show()
```
