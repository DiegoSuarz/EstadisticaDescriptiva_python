```python
import matplotlib.pyplot as plt
import numpy as np
from sklearn import datasets, linear_model
from sklearn.metrics import mean_squared_error, r2_score
```

```python
diabetes = datasets.load_diabetes()
diabetes
```

```python
type(diabetes)
```

```python
type(diabetes.data)
```

```python
diabetes.data.shape
```

```python
x = diabetes.data[:, np.newaxis,2] # solo me quedo con la primera columna para la variable independiente
x
```

```python
y = diabetes.target #variable dependinte
```

```python
x_train = x[:-60] # todas las filas menos 60 filas
x_test = x[-60:]

y_train = y[:-60]
y_test = y[-60:]
```

```python
modelo = linear_model.LinearRegression()
modelo.fit(x_train ,y_train)
```

```python
# coeficientes:
modelo.coef_
```

```python
# intercepto del modelo:
modelo.intercept_
```

$$\tilde{y} = 940.05574083\cdot x + 152.80161883617293 $$ 

```python
plt.scatter(x, y, color = 'black')
plt.plot(x_train, modelo.predict(x_train), color = 'blue', linewidth = 3)
plt.show()
```

$$ MSE = \frac{\sum_{i=1}^n(y_i - \tilde{y_i})^2}{n}

```python
mean_squared_error(y, modelo.predict(x))
```

```python
r2_score(y, modelo.predict(x))
```

0.3438149143893624 no se acerca al valor optimo de 0.9, asi que podemos afirmar que no tiene una relacion lineal

## Testing:

```python
y_pred = modelo.predict(x_test)

mean_squared_error(y_test, y_pred)
```

```python
r2_score(y_test, y_pred)
```
