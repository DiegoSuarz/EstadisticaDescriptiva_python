```python
import numpy as np
import pandas as pd
import seaborn as sns

titanic = sns.load_dataset('titanic')   
titanic.head()
```

```python
titanic.info()
```

```python
titanic.describe()
```

```python
# convertir variables int a categoricas
titanic['survived'] = pd.Categorical(titanic['survived'])
titanic['pclass'] = pd.Categorical(titanic['pclass'])
titanic['sex'] = pd.Categorical(titanic['sex'])
titanic['deck'] = pd.Categorical(titanic['deck'])
```

## Tablas de contingencia de una variable:

```python
tab = pd.crosstab(index = titanic['survived'], # tabla de frecuencias de sobrevivientes y clase de pasajeros   
                  columns = 'pasajeros')
type(tab)
tab
```

```python
pd.crosstab(index = titanic['sex'],
           columns = "pasajeros")
```

```python
tab_c = pd.crosstab(index = titanic['deck'], # tabla de frecuencias de cubiertas
            columns= "count")
tab_c
```

```python
tab_c.sum()
```

```python
tab_c.iloc[1:5]
```

```python
# tabla de frecuencias relativas
tab_c/tab_c.sum()
```

## Tabla de contingencia para 2 variables:

```python
survived_sex = pd.crosstab(index = titanic['survived'],
                           columns = titanic['sex'])

survived_sex.index = ['died', 'survived'] # cambiar los indices de la tabla de frecuencias de sobrevivientes y sexo
survived_sex
```

```python
survided_class = pd.crosstab(index = titanic['survived'],
                                columns = titanic['pclass'],
                                margins = True) # margenes de la tabla de frecuencias de sobrevivientes y clase de pasajeros
survided_class.index = ['murio', 'sobrevivio', 'total_clase']
survided_class.columns = ['primera', 'segunda', 'tercera', 'total_sobrevivio']
survided_class
```

## Frecuencias relativas globales:

```python
survided_class/survided_class.loc["total_clase", "total_sobrevivio"] # tabla de frecuencias relativas de sobrevivientes y clase de pasajeros
```

## Frecuencias relativa marginales

```python
# primera forma de hacer la tabla de frecuencias relativas marginales por columna
survided_class/survided_class.loc["total_clase"] # tabla de frecuencias relativas marginales por columna
```

```python
# segunda forma de hacer la tabla de frecuencias relativas marginales por columna
survided_class.div(survided_class.loc["total_clase"], axis= 1) # tabla de frecuencias relativas marginales por columna
```

```python
#primera forma:
(survided_class.T/survided_class["total_sobrevivio"]).T # tabla de frecuencias relativas marginales por fila
```

```python
#segunda forma:
survided_class.div(survided_class["total_sobrevivio"], axis = 0) # tabla de frecuencias relativas marginales por fila
```

## Tablas multidimensionales:

```python
surv_sex_class = pd.crosstab(index = titanic['survived'],
                            columns=[titanic["sex"], titanic["pclass"]],
                            margins = True 
                            )
surv_sex_class
```

```python
surv_sex_class["female"] # tabla de frecuencias de sobrevivientes y clase de pasajeros para mujeres
```

```python
surv_sex_class["female"][1] # tabla de frecuencias de sobrevivientes y clase de pasajeros para mujeres en primera clase
```

```python
# Tabla de frecuencias relativas Global
# primera forma
surv_sex_class/surv_sex_class.loc["All", ["All"][0]][0]
```

```python
# segunda forma:
surv_sex_class/surv_sex_class.loc["All", "All"][0]
```

```python

surv_sex_class/surv_sex_class.loc["All"] #Tabla de frecuencias marginales de sobrevivientes y clase de pasajeros para mujeres por columna
```

```python
# Otra forma de las tablas de frecuencias:

#Tabla de frecuencias globales totales:
surv_sex_class = pd.crosstab(index = titanic['survived'],
                            columns=[titanic["sex"], titanic["pclass"]],
                            margins = True, 
                            normalize= True,
                            rownames= ["Survived"],
                            colnames=["sex","clase"] 
                            )
surv_sex_class
```

```python
#Tabla de frecuencias marginales por fila:
surv_sex_class = pd.crosstab(index = titanic['survived'],
                            columns=[titanic["sex"], titanic["pclass"]],
                            margins = True, 
                            normalize= "index",
                            rownames= ["Survived"],
                            colnames=["sex","clase"] 
                            )
surv_sex_class
```

```python
#Tabla de frecuencias marginales por columna:
surv_sex_class = pd.crosstab(index = titanic['survived'],
                            columns=[titanic["sex"], titanic["pclass"]],
                            margins = True, 
                            normalize= "columns",
                            rownames= ["Survived"],
                            colnames=["sex","clase"] 
                            )
surv_sex_class
```

```python
# Otro ejemplo de tabla de frecuencias
#tabla normal
pd.crosstab(index = titanic['survived'],
                                columns = titanic['pclass'],
                                margins = True,
                            ) # margenes de la tabla de frecuencias de sobrevivientes y clase de pasajeros
```

```python
#global
pd.crosstab(index = titanic['survived'],
                                columns = titanic['pclass'],
                                margins = True,
                                normalize=True
                                ) # margenes de la tabla de frecuencias de sobrevivientes y clase de pasajeros
```

```python
#marginal por fila
pd.crosstab(index = titanic['survived'],
                                columns = titanic['pclass'],
                                margins = True,
                                normalize="index"
                                ) # margenes de la tabla de frecuencias de sobrevivientes y clase de pasajeros
```

```python
#marginal por columna
pd.crosstab(index = titanic['survived'],
                                columns = titanic['pclass'],
                                margins = True,
                                normalize="columns"
                                ) # margenes de la tabla de frecuencias de sobrevivientes y clase de pasajeros
```
