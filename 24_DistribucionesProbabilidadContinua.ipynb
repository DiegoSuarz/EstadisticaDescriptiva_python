## Ejemplo de distribucion uniforme:

Supongamos que $X\sim U([0,1])$ entonces podemos estudiar sus par√°metros:

```python
from scipy.stats import uniform
import matplotlib.pyplot as plt
import numpy as np

a = 0
b = 1

loc = a
scale = b - a

mean, var, skew, kurt = uniform.stats(moments = 'mvsk')

print("Media %f"%mean)
print("Varianza %f"%var)
print("Sesgo %f "%skew)
print("Curtosis %f"%kurt)
```

```python
fig, ax = plt.subplots(1,1)

rv = uniform(loc = loc, scale = scale)
x = np.linspace(-0.1, 1.1, 120)
ax.plot(x, rv.pdf(x), 'k-', lw = 2, label = 'U(0,1)')
ax.legend(loc = 'best', frameon = False)
plt.show()
```

```python
fig, ax = plt.subplots(1,1)

rv = uniform(loc = loc, scale = scale)
x = np.linspace(-0.1, 1.1, 120)
ax.plot(x, rv.pdf(x), 'k-', lw = 2, label = 'U(0,1)')

r = rv.rvs(size = 1000000)
ax.hist(r, density= True, histtype= "stepfilled", alpha = 0.25)
ax.legend(loc = 'best', frameon = False)
plt.show()
```

## Distribucion Exponencial:

```python
from scipy.stats import expon
import numpy as np
import matplotlib.pyplot as plt

lam = 3
rv = expon(scale = 1/lam)

mean, var, skew, kurt = rv.stats(moments= 'mvsk')
print("Media %f"%mean)
print("Varianza %f"%var)
print("Sesgo %f "%skew)
print("Curtosis %f"%kurt)
```

```python
x = np.linspace(0, 2, 1000)

fig, ax = plt.subplots(1,1)
ax.plot(x, rv.pdf(x), 'r-', lw = 5, alpha = 0.6, label = "Exp(10)")

r = rv.rvs(size = 100000)
ax.hist(r, density=True,bins= 50, histtype= 'stepfilled', alpha = 0.2)

plt.show()
```
