## Manipulacion de datos con Pandas:

```python
df3 = pd.DataFrame({
    'key' : ['A', 'B', 'C', 'A', 'B', 'C'],
    'value1' : range(6),
    'value2' : np.random.randint(0, 100, 6)
})
df3
```

```python
df3.groupby('key').aggregate(['min', np.median, np.mean, max])
```

```python
df3.groupby('key').aggregate({
    'value1' : 'min',
    'value2' : 'max'
})
```

## FILTER

```python
df3.groupby('key').std()
```

```python
def filter_func(x):
    return x['value2'].std() > 5 # Desviacion estandar mayor a 5

display(df3.groupby('key').filter(filter_func))
```

## TRANSFORM

```python
df3.groupby('key').transform(lambda x: x - x.mean())
```

## APPLY

```python
def norm_by_col2(x):
    x['value1'] /= x['value2'].sum()
    return x

df3.groupby('key').apply(norm_by_col2)
```

## Gráficos desde data frames y exportación:

```python
year = [year for year in range(1900, 2020)]
deads = [(y + np.random.uniform(0,100)- 1850) for y in year]

df5 = pd.DataFrame({
    'year' : year,
    'deads' : deads
})
df5
```

```python
df5.plot(x= 'year', y = 'deads', figsize=(10, 8))
```

```python
df5.plot(kind = 'scatter',x= 'year', y = 'deads', figsize=(10, 8))
```

```python
my_plot = df5.plot(x= 'year', y = 'deads', figsize=(10, 8))
my_fig = my_plot.get_figure()
my_fig.savefig("archivos exportados/grafico.png")
```
