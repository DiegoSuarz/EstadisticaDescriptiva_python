```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from plotnine.data import diamonds

plt.style.use("ggplot")
```

```python
diamonds.shape
```

```python
diamonds.head(10)
```

```python
## Histograma:
# Peso en kilates:
diamonds.hist(column='carat', figsize= (8,8), color = 'blue',
               bins = 50, range = (0, 3.5))
```

```python
## Filtro:
diamonds[diamonds["carat"] > 3.5] #diaomns que pareces outliers
```

```python
## Boxplots:
diamonds.boxplot(column= "carat", figsize=(8,8))
plt.show()
```

```python
diamonds.boxplot(column= "carat",by = "clarity", figsize=(8,8))
```

```python
diamonds.boxplot(column= "price",by = "clarity", figsize=(8,8))
```

```python
diamonds.boxplot(column= "carat",by = "color", figsize=(8,8))
```

```python
## Densidades:
plt.clf() # limpiar la figura:
diamonds["carat"].plot(kind = "density", figsize = (8,8), xlim=(0,5))
plt.show()
```

```python
# Tabla de frecuencias de la variable clarity:
carat_table = pd.crosstab(index=diamonds['clarity'], columns='count')
carat_table
```

```python
# Barplot del punto anterior:
carat_table.plot(kind='bar', figsize=(8,8))
plt.show()
```

```python
carat_table_2 = pd.crosstab(index=diamonds["clarity"],
                            columns=diamonds["color"])
carat_table_2
```

```python
carat_table_2.plot(kind='bar', 
                    figsize=(8,8),
                    stacked=True # Apilar las barras (una sobre otra)
                    )
plt.show()
```

```python
carat_table_2.plot(kind='bar', 
                    figsize=(8,8),
                    stacked=False # Poner las barras una al lado de otra
                    )
plt.show()
```

```python
# Scatterplot:
diamonds.plot(kind = "scatter", x = "carat", y = "price",
               figsize = (10,10), 
               ylim = (0, 20000),
               xlim = (0, 6),
               alpha = 0.3
               )
plt.show()
```
